# Move Videos

```python
import shutil
video_name = "Luke Belmar Visits Andrew & Tristan Tate"
video_id = "ZykiUbwHaYg"
```

```python
shorts_path = os.path.join(Utils.General.BASE_PATH,video_id,"shorts")
videos_folder = r"C:\Users\BEN\Desktop\Content\Short Videos Automation\TateTok\Videos\TikTok"
shorts_names = [filename for filename in os.listdir(shorts_path) if filename.endswith(".mp4")]
for short_name in shorts_names:
    new_name = video_name + " #" + short_name
    old_path = os.path.join(shorts_path,short_name)
    new_path = os.path.join(videos_folder,new_name)
    shutil.copyfile(old_path, new_path)
```
